# Creador de Tarjetas

Este programa tiene como objetivo crear las tarjetas de preguntas de un juego de mesa. Las preguntas están divididas en 11 archivos csv (coma separated values), cada uno corresponde a una categoría diferente.

Cada archivo tiene 4 columnas:

1. Preguntas
2. Respuestas
3. Dificultad (1-3)
4. Subcategoría

## La idea:

Este programa debe elegir las preguntas que iran en cada tarjeta, las tarjetas deben tener 5 preguntas cuyas categorías no deben repetirse. Las categorías de cada tarjeta deben ser elegidas al azar, al igual que las preguntas. Ninguna tarjeta debe tener una pregunta que esté en otra tarjeta, además cada tarjeta debe tener un número de seguimiento asociado y su dificultad general calculada.

### 1) Importar librerías

```python
import numpy as np
import pandas as pd
import os
import random
```

### 4) Definimos las funciones que se utilizarán en el programa

```python
def random_exclude(range_start, range_end, excludes): #Obtiene un número entero al azar en un rango excluyendo ciertos números
    r = random.randint(range_start, range_end)
    if r in excludes: #Si el número obtenido calza con una exclusión se llama a la función de nuevo.
        return random_exclude(range_start, range_end, excludes)
    return r
```

### 7) Ahora creamos cada tarjeta y la guardamos en un archivo separado

Las tarjetas quedarán guardadas en la carpeta "Tarjetas" y cada una tendrá su propio archivo llamado "Tarjeta_Número_#.csv"
Además se agregará la columna que entrega el valor de dificultad de la tarjeta llamada "Valor Tarjeta", la cual entrega un valor textual que puede ser:
1. Fácil
2. Intermedia
3. Difícil
y entre paréntesis el valor de dificultad numérico exácto entregado por
\begin{equation}
\frac{\sum_{i=1}^{5} d_{i}}{5}
\end{equation}
Donde $d_{i}$ es la dificultad de la i-ésima pregunta

```python
i = 0 #Contador de tarjetas, va de 1 en 1
k= 0 #Contador de preguntas, va de 5 en 5
Nombres =['Tarjetas/Tarjeta_Número_'+str(j) for j in range(Cantidad_Tarjetas)]
for i, card_csv_name in enumerate(Nombres):
    Tarjeta = Todas_las_tarjetas.iloc[k:k+5]
    k=k+5
    dif_temp=(Tarjeta.iloc[:,2].sum(axis=0))/5
    if dif_temp <=1.7: #Asigna dificultad a la tarjeta,
        dif = "Fácil"
    elif dif_temp > 1.7 and dif_temp <= 2.4:
        dif = "Intermedia"
    else:
        dif = "Difícil"  
    dificultad = [dif + '('+str(dif_temp)+')',dif + '('+str(dif_temp)+')',
                  dif + '('+str(dif_temp)+')',dif + '('+str(dif_temp)+')',dif + '('+str(dif_temp)+')']    
    Tarjeta['Valor Tarjeta'] = dificultad
    Tarjeta.to_csv(card_csv_name+'.csv',index=False)
```
